```python
x_ones=torch.ones_like(x_data)
print(f"Ones Tensor: \n {x_ones}\n")
```

```python
x_rand=torch.rand_like(x_data,dtype=torch.float)
print(f"Random Tensor: \n {x_rand}\n")
```

```python
shape=(2,3)
rand_tensor=torch.rand(shape)
ones_tensor=torch.ones(shape)
zero_tensor=torch.zeros(shape)
print(f"Random Tensor:\n{rand_tensor}\n")
print(f"Ones Tensor:\n{ones_tensor}\n")
print(f"Zero Tensor:\n{zero_tensor}\n")
```

```python
tensor=torch.rand(3,4)
print(f"datatype of tensor {tensor.dtype}")
print(f"device tensor is stored on: {tensor.device}")
tensor=tensor.cuda()
print(f"device tensor is stored on: {tensor.device}")
```

```python
tensor=torch.ones(4,4)
print(f"First row is : {tensor[0]}")
print(f"First colum is : {tensor[:,0]}")
print(f"Last colum is : {tensor[:,-1]}")
tensor[:,-1]=0
print(tensor)
```

```python
t1=torch.cat([tensor,tensor,tensor],dim=1)
print(t1)
print(t1.size())
```

```python
t2=torch.stack([tensor,tensor,tensor],dim=1)
print(f"Original tensor size :{tensor.size()}")
print(f"tensor stack in 1 dimension: {t2.size()}")
t2=torch.stack([tensor,tensor,tensor],dim=0)
# usually use this to stack the training data
print(f"tensor stack in 0 dimension: {t2.size()}")
```

```python
y1=tensor@tensor.T
y2=tensor.matmul(tensor.T)
y3=torch.rand_like(y1)
torch.matmul(tensor,tensor.T,out=y3)
print(y3)
```

```python
z1=tensor*tensor
z2=tensor.mul(tensor)
z3=torch.rand_like(z1)
torch.mul(tensor,tensor.T,out=z3)
print(z3)
```

```python
agg=tensor.sum()
print(agg)
agg.item()
```

```python
#using the _ to operation the tensor in-place
print(f"{tensor}\n")
tensor.add_(5)
print(tensor)
```
